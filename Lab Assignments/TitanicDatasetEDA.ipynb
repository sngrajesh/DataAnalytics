```python
# Function to detect and impute outliers using IQR method
def detect_and_impute_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Replace outliers with median
  df[column] = np.where(df[column] < lower_bound, df[column].median(), df[column])
  df[column] = np.where(df[column] > upper_bound, df[column].median(), df[column])

  return df

# Iterate through numerical columns and perform outlier detection & imputation
for column in df_fork.select_dtypes(include=np.number):
  df_fork = detect_and_impute_outliers_iqr(df_fork, column)
```
