17. Write a Pandas program to get the items which are not common of two given series.

Sample Output:  

Original Series: sr1:

|   | Value |
|---|-------|
| 0 |   1   |
| 1 |   2   |
| 2 |   3   |
| 3 |   4   |
| 4 |   5   |

dtype: int64

sr2:

|   | Value |
|---|-------|
| 0 |   2   |
| 1 |   4   |
| 2 |   6   |
| 3 |   8   |
| 4 |  10   |

dtype: int64

Items of a given series not present in another given series:

|   | Value |
|---|-------|
| 0 |   1   |
| 1 |   3   |
| 2 |   5   |
| 3 |   6   |
| 4 |   8   |
| 5 |  10   |

dtype: int64

````python
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])
print(sr1)
print(sr2)


# Method 1:- (Union - Intersection)
sr3 = pd.Series(np.union1d(sr1, sr2))
sr4 = pd.Series(np.intersect1d(sr1, sr2))
print(sr3[~sr3.isin(sr4)])

# Method 2:- concat method
print(pd.concat([sr1[~sr1.isin(sr2)], sr2[~sr2.isin(sr1)]]))

# Method 3:- set operation with xor
print(pd.Series(list(set(sr1) ^ set(sr2)), name="Value"))
````

````python
s = pd.Series([1, 8, 7, 5, 6, 5, 3, 4, 7, 1])

left_greater = s.shift(1) < s
# N, 1, 8, 7, 5, 6, 5, 3, 4, 7
# 1, 8, 7, 5, 6, 5, 3, 4, 7, 1
# F, T, F, F, T, F, F, T, T, F

right_greater = s.shift(-1) < s
# 1, 8, 7, 5, 6, 5, 3, 4, 7, N
# 1, 8, 7, 5, 6, 5, 3, 4, 7, 1
# F, T, T, F, T, T, F, F, T, F

print('Positions of the values surrounded by smaller values on both sides: ' ,s[left_greater & right_greater].index.tolist())
````
