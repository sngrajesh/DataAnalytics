1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.



```python
# @title{vertical-output:true}
df['Date']=pd.to_datetime(df['Date'])
```

```python
# @title{vertical-output:true}
plt.figure()
# fdf = df[(df['Date'] >= '2020-01-01') & (df['Date'] <= '2020-12-31')]
fdf=df[df["Date"].between("2020-04-06", "2020-09-24")]
plt.plot(fdf['Date'],fdf['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Alphabet Inc. Stock Price (2020)')
plt.grid(True)
plt.show()
```

12. Write a Pandas program to create a plot of stock price and trading volume of Alphabet Inc. between two specific dates.



```python
# @title{vertical-output:true}
plt.figure()
df2 = df[df["Date"].between("2020-04-06", "2020-04-18")][['Date','Close','Volume']]
df2.set_index(['Date'], inplace=True)
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["Close"])
plt.title('Historical stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
bottom_plt = plt.subplot2grid((5,4), (3,0), rowspan=1, colspan=4)
bottom_plt.bar(df2.index, df2['Volume'])
plt.title('\nAlphabet Inc. Trading Volume', y=-0.60)
plt.gcf().set_size_inches(12,8)
```

13. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of Alphabet Inc. between two specific dates.



```python
# @title{vertical-output:true}
plt.figure()
df2 = df[df["Date"].between("2020-04-06", "2020-04-18")][['Date','Open','High','Low','Close','Adj Close','Volume']]
df2.set_index(['Date'], inplace=True)

top_plt = plt.subplot2grid((20,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["Open"])
plt.title('Open stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
top_plt = plt.subplot2grid((20,4), (3, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["High"])
plt.title('High stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
top_plt = plt.subplot2grid((20,4), (6, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["Low"])
plt.title('Low stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
top_plt = plt.subplot2grid((20,4), (9, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["Close"])
plt.title('Close stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
top_plt = plt.subplot2grid((20,4), (12, 0), rowspan=3, colspan=4)
top_plt.plot(df2.index, df2["Adj Close"])
plt.title('Adj Close stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
bottom_plt = plt.subplot2grid((20,4), (15,0), rowspan=1, colspan=4)
bottom_plt.bar(df2.index, df2['Volume'])
plt.title('\nAlphabet Inc. Trading Volume', y=-0.60)
plt.gcf().set_size_inches(12,8)
```

```python
# @title{vertical-output:true}
plt.figure()
df2 = df[df["Date"].between("2020-04-06", "2020-04-18")][['Date','Open','High','Low','Close','Adj Close','Volume']]
df2.set_index(['Date'], inplace=True)

top_plt = plt.subplot2grid((20,4), (0, 0), rowspan=15, colspan=4)
top_plt.plot(df2['Open'], label='Open')
top_plt.plot(df2['High'], label='High')
top_plt.plot(df2['Low'], label='Low')
top_plt.plot(df2['Close'], label='Close')
plt.title('Stock prices of Alphabet Inc. [2020-04-06, 2020-04-18]')
plt.legend(loc='best')

bottom_plt = plt.subplot2grid((20,4), (17,0), rowspan=3, colspan=4)
bottom_plt.bar(df2.index, df2['Volume'])
plt.title('Alphabet Inc. Trading Volume', y=-0.60)

plt.gcf().set_size_inches(12, 8)
plt.tight_layout()
plt.show()
```

14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.



```python
# @title{vertical-output:true}
# Extract and isolate 'Close' column, converting to DataFrame
df1 = df['Close'].to_frame()

# Calculate Cumulative Moving Average (CMA) with a window of 30
df1['CMA30'] = df1['Close'].expanding(30).mean()


# Print DataFrame
df1.plot()
plt.show()
```

15. Write a Pandas program to create a plot of adjusted closing prices, 30 days simple moving average and exponential moving average of Alphabet Inc. between two specific dates.



```python
# @title{vertical-output:true}
# Extract and isolate 'Close' column, converting to DataFrame
df1 = df['Close'].to_frame()

# Calculate Cumulative Moving Average (CMA) with a window of 30
df1['CMA30'] = df1['Close'].expanding(30).mean()

# Calculate Exponential Moving Average (EWMA) with a span of 30
df1['EWMA30'] = df1['Close'].ewm(span=30).mean()

# Print DataFrame
df1.plot()
plt.show()
```

16. Write a Pandas program to create a scatter plot of the trading volume/stock prices of Alphabet Inc. stock between two specific dates.



```python
# @title{vertical-output:true}
import matplotlib.pyplot as plt

start_date = '2020-04-06'
end_date = '2020-05-18'

filtered_df = df[df['Date'].between(start_date, end_date)]

plt.figure(figsize=(10, 6))

plt.scatter(filtered_df['Date'], filtered_df['Volume'], alpha=0.5)
plt.scatter(filtered_df['Date'], filtered_df['Close'], alpha=0.5)

plt.title('Alphabet Inc. Stock: Trading Volume vs. Stock Price')
plt.xlabel('Stock Price (Close)')
plt.ylabel('Trading Volume')
plt.grid(True)
plt.show()
```
