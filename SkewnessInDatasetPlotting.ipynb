```python
filepaths = [{
        'path':'/content/drive/MyDrive/airline_dec_2008_50k.csv',
        'field': 'Distance'
    },
    {
        'path': '/content/drive/MyDrive/airline_passenger_satisfaction.csv',
        'field': 'Flight Distance'
    },
    {
        'path':'/content/drive/MyDrive/airline_dec_2008_50k.csv',
        'field': 'CRSDepTime'
    }
]
```

```python
# @title {vertical-output:true}
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_data(filepath):
    airline_df = pd.read_csv(filepath['path'])
    print(airline_df.columns)
    distance = airline_df[filepath['field']]

    mean_distance = distance.mean()
    median_distance = distance.median()
    skewness_distance = distance.skew()

    plt.figure(figsize=(8,6))
    sns.histplot(distance, bins=20, color='skyblue', kde=True)
    plt.axvline(mean_distance, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_distance:.2f}')
    plt.axvline(median_distance, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_distance:.2f}')
    plt.title(f'Distribution of {filepath["field"]} / Skewness: {skewness_distance:.2f}')
    plt.legend()
    plt.xlabel(filepath['field'])
    plt.ylabel('Frequency')
    plt.show()
```

```python
# @title {vertical-output:true}
for filepath in filepaths:
    plot_data(filepath)
```
