#### Question 3 : FInd the probabiliy that customers gives tips more than 15%

```python
tips_df = pd.read_csv('/content/drive/MyDrive/tips.csv')
```

```python
# Binary outcome : succes if (tip / total_bill) > 0.15, otherwise failure

def classify_tile(row):
  if (row['tip'] / row['total_bill']) > 0.15:
    return 1 # Success
  else:
    return 0  # Failure


# Apply function to create the binarry column
tips_df['tips_binary'] = tips_df.apply(classify_tile, axis=1)
```

```python
# Total nuimber of trails (total obersvation)
n = len(tips_df)

# Calculate the probability of success (tips / total_bill > 0.15)
k = tips_df['tips_binary'].sum()

# Probability of success
p = k / n
```

```python
# Define numberof trails for the binimial distribution
trails = 100

# Generate  values (number of successes form 0 to 100)
x = np.arange(0,trails+1)
```

```python
# Calculate the PMF of each number of success
pmf_values  = binom.pmf(x,trails,p)
```

```python
# Calculate the CDF of each number of success
cdf_values = binom.cdf(x,trails,p)
```

```python
#@title{vertical-output:true}

# cumulative_probability = 0;

for i in x:
    # cumulative_probability = binom.cdf(i,trails,p)
    # print(f"Probability of {i} successes: {pmf_values[i]:.6f}...Cumulative Probability : {cumulative_probability:.4f}")
    print(f"Probability of {i} successes: {pmf_values[i]:.6f}...Cumulative Probability : {cdf_values[i]:.4f}")
```

```python
#@title{vertical-output:true}

# Plotting the pmf all pmf
plt.figure(figsize=(10,6))
plt.plot(pmf_values)
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass Function')
plt.title('Probability Mass Function of Binomial Distribution')
plt.show()
```

```python
#@title{vertical-output:true}

# Plotting the cdf
plt.figure(figsize=(10,6))
plt.point(cdf_values)
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass Function')
plt.title('Probability Mass Function of Binomial Distribution')
plt.show()
```
