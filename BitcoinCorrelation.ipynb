```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SP500 oil gold bitcoin.csv')
print(df.head())
```

```python
# correlation
corr_bitcoin_gold = df[['BITCOIN', 'Gold']].corr()
corr_bitcoin_SP500 = df[['BITCOIN', 'S&P500']].corr()
print("Correlation between bitcoin and gold:\n ", corr_bitcoin_gold)
print("Correlation between bitcoin and SP500:\n ", corr_bitcoin_SP500)
```

```python
# covariance
cov_bitcoin_gold = df[['BITCOIN', 'Gold']].cov()
cov_bitcoin_SP500 = df[['BITCOIN', 'S&P500']].cov()
print("Covariance between bitcoin and gold: \n", cov_bitcoin_gold)
print("Covariance between bitcoin and SP500: \n", cov_bitcoin_gold)
```
