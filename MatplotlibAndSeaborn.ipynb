### Matplotliib OOP:
#### Create a figure object and use methods/attributes
Bettert when dealing with a canvas that has multiple plots on it

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
a = np.linspace(0, 10, 11)
b = a**4
print(a)
print(b)
````

````python
x = np.arange(0,10)
y=2*x
print(x)
print(y)
````

````python
# Create an empty canvas
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height (rane 0 to 1)
axes.plot(x,y)
plt.show()
````

````python
# Create an empty canvas
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height (rane 0 to 1)
axes.plot(a,b)
plt.show()
````

Subplot

````python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Parameters at the axes level
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(y,x)
axes[1][1].plot(b,a)



# Use left, right, top, bottom to stretch subplots
# Use wspace, hspace to add spacing between subplots
fig.subplots_adjust(
    left=0.1,
    bottom=0.1,
    right=0.9,
    top=0.9,
    wspace=0.5,
    hspace=0.5
)

plt.show()
````

## Seaborn

````python
import seaborn as sns
````

````python
# Generate random data
np.random.seed(101)
x = np.random.rand(100)
y = np.random.rand(100)
print(x)
print(y)
````

````python
# Create scatter plot using matplotlib
plt.figure(figsize=(10,10))
plt.scatter(x,y, color='blue', alpha=0.6)
plt.title('Scatter Plot using Matplotlib', fontsize=15)
plt.xlabel('X-axis',fontsize=12)
plt.ylabel('Y-axis',fontsize=12)
plt.grid(True)
plt.show()
````

````python
# Create scatter plot using seaborn
sns.set(style='whitegrid')
plt.figure(figsize=(10,10))
sns.scatterplot(x=x,y=y)
plt.title('Scatter Plot using Seaborn', fontsize=15)
plt.xlabel('X-axis',fontsize=12)
plt.ylabel('Y-axis',fontsize=12)
plt.grid(True)
plt.show()
````

## Plot Using DataFrame

````python
import pandas as pd
````

````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/heart.csv')
print(df.shape)
print(df.head(3))
print(df.info())
````

````python
ages = df.groupby('age').median().reset_index()
print(ages.shape)
print(ages.head(10))
````

````python
# Simple pandas plot
ages.plot.line('age','chol')
plt.show()
````

````python
# Matplotlib
fig, ax = plt.subplots()
ax.plot(ages['age'], ages['chol'], ls=":",lw=1.7)
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')
ax.set_title('Cholesterol vs Age')
plt.show()
````

````python
# Seaborn
sns.lineplot(x='age',y='chol',data=ages, linestyle=":", linewidth=1.7)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Cholesterol vs Age')
plt.show()
````

````python
# Simple pandas plot
ages.plot.line('age','thalach')
plt.show()
````

````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dm_office_sales.csv')
print(df.shape)
print(df.info())
print(df.head(3))
print(df.columns)
````

````python
plt.figure(figsize=(4,4), dpi = 200)  ## dpi - dots per inch of space
sns.countplot(x='division', data=df)
plt.xticks(rotation=30)
plt.show()
````

````python
plt.figure(figsize=(4,4), dpi = 200)
sns.countplot(x='level of education', data=df)
plt.xticks(rotation=30)
plt.show()
````

````python
 # Breakdown within another category with 'hue'
plt.figure(figsize=(4,4), dpi = 200)
sns.countplot(x='level of education', data=df, hue='training level')
plt.xticks(rotation=30)
plt.show()
````

````python
# Using palette
plt.figure(figsize=(4,4), dpi = 200)
sns.countplot(x='level of education', data=df, hue='training level', palette='Set1')
plt.xticks(rotation=30)
plt.show()
````

| Plot Type        | Use Case Example                                                                                  | Code Example                                                                                                                                         |
|------------------|---------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------|
| Line Plot        | Used for visualizing trends or changes over time.                                                  | ```import matplotlib.pyplot as plt  plt.plot([1, 2, 3, 4], [10, 20, 25, 30])  plt.show()```                                                           |
| Bar Plot         | Used for comparing different categories.                                                           | ```import seaborn as sns sns.barplot(x=["A", "B", "C"], y=[10, 20, 15]) plt.show()```                                                                 |
| Histogram        | Used for showing the distribution of a dataset.                                                    | ```import matplotlib.pyplot as plt  plt.hist([1, 2, 2, 3, 4, 4, 4, 5])  plt.show()```                                                                |
| Scatter Plot     | Used for showing the relationship between two continuous variables.                                | ```import matplotlib.pyplot as plt  plt.scatter([1, 2, 3, 4], [10, 15, 13, 20])  plt.show()```                                                        |
| Box Plot         | Used for visualizing the spread and distribution of a dataset.                                     | ```import seaborn as sns sns.boxplot(data=[10, 20, 15, 30, 25]) plt.show()```                                                                         |
| Heatmap          | Used for visualizing matrix-like data or correlation matrices.                                     | ```import seaborn as sns  data = [[1, 2], [3, 4]]  sns.heatmap(data, annot=True)  plt.show()```                                                       |
| Violin Plot      | Used for visualizing the distribution of data across different categories with KDE information.    | ```import seaborn as sns sns.violinplot(data=[10, 20, 15, 30, 25]) plt.show()```                                                                      |
| Pie Chart        | Used for showing proportions of categories as parts of a whole.                                    | ```import matplotlib.pyplot as plt  plt.pie([10, 20, 30], labels=["A", "B", "C"], autopct="%1.1f%%")  plt.show()```                                   |
| Pair Plot        | Used for visualizing relationships between multiple pairwise features in a dataset.                | ```import seaborn as sns  sns.pairplot(sns.load_dataset("iris"))  plt.show()```                                                                       |
| Count Plot       | Used for visualizing the frequency of occurrences of categorical data.                             | ```import seaborn as sns sns.countplot(x=["A", "B", "A", "C", "B", "A"]) plt.show()```                                                                |
| KDE Plot         | Used for estimating and plotting the probability density function of a dataset.                    | ```import seaborn as sns  sns.kdeplot(data=[10, 20, 15, 25], fill=True) plt.show()```                                                                 |
| Joint Plot       | Used for showing the relationship between two variables along with their individual distributions. | ```import seaborn as sns  sns.jointplot(x=[1, 2, 3], y=[4, 5, 6])  plt.show()```                                                                      |
| Swarm Plot       | Used for visualizing the distribution of data points while avoiding overlap.                       | ```import seaborn as sns sns.swarmplot(x=["A", "B", "C"], y=[10, 20, 15]) plt.show()```                                                               |
| Area Plot        | Used for visualizing cumulative totals over time or comparing different series.                    | ```import matplotlib.pyplot as plt  plt.fill_between([1, 2, 3, 4], [10, 15, 13, 20])  plt.show()```                                                   |
