```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Sets/airline_passenger_satisfaction.csv')
```

```python
male_df = df[df['Gender']=='Male']
male_df.columns
```

```python
# Analyze On-bard Service and class (Independent Test - Male Passanger)
category1 = "Class"
category2 = "On-board Service"
```

```python
# Create contigency table
crosstab = pd.crosstab(male_df[category1], male_df[category2])
```

```python
# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(crosstab)

# Print the results
print(f"Category1 : {category1}, Category2 : {category2}")
print('Crosstab', crosstab)
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python
if p < 0.05:
    print("Reject null hypothesis: There is a significant association between the categories.")
else:
    print("Failed to reject null hypothesis: There is no significant association between the categories.")
```
