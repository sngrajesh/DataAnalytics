# Imports

```python
import pandas as pd
from scipy import stats
from math import sqrt
from scipy.stats import ttest_rel
from scipy.stats import t
```

# One sample t-test

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Sets/Mall_Customers.csv')
```

```python
age_data = df['Age']
```

```python
# Hypothesized population average
pop_avg_age = 40
```

```python
# Perform one sample t-test
t_statistic, p_value = stats.ttest_1samp(age_data, pop_avg_age)
degrees_of_freedom = len(age_data) - 1

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
print("Degrees of freedom:", degrees_of_freedom)
```

```python
# Interpretation
alpha = 0.025
if p_value < alpha:
    print("Reject the null hypothesis (MEAN AGE = 40). There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis (MEAN AGE = 40). There is no significant difference between the sample mean and the population mean.")
```

```python
critical_value = stats.t.ppf(1 - alpha, degrees_of_freedom)
print("Critical value:", critical_value)
```

```python
# Two-tailed t-test (Because we are just checking if sample mean = population)
# So, we need to take absolute value of t-statistics
if (abs(t_statistic) > critical_value):
    print("Reject the null hypothesis (MEAN AGE = 40).")
else:
    print("Fail to reject the null hypothesis (MEAN AGE = 40).")
```

# Tow sample t-test

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Sets/Mall_Customers.csv')
```

```python
# Seperate data by genre
male_spending = df[df['Genre'] == 'Male']['Spending Score (1-100)']
female_spending = df[df['Genre'] == 'Female']['Spending Score (1-100)']
```

```python
# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(male_spending, female_spending)

n_male  = len(male_spending)
n_female = len(female_spending)
degrees_of_freedom = n_male + n_female - 2

alpha = 0.025

# Find Critical value
critical_value = stats.t.ppf(1 - alpha, degrees_of_freedom)

# Print the result
print("T-statistic:", t_statistic)
print("P-value:", p_value)
print("Degrees of freedom:", degrees_of_freedom)
print("Critical value:", critical_value)
```

```python
if p_value < alpha:
    print("There is a significant difference in spending scores between male and female group.")
else:
    print("There is no significant difference in spending scores between male and female group")
```

```python
if abs(t_statistic) > critical_value:
    print("There is a significant difference in spending scores between male and female group.")
else:
    print("There is no significant difference in spending scores between male and female group")
```

# Paired sample t-test

```python
# Read the CSV data
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data Sets/cust_seg.csv")
```

```python

# Select relevant columns (assuming pre_usage and Post_usage_1month are numeric)
usage_data = df[["pre_usage", "Post_usage_1month"]]

```

```python
# Perform paired t-test
tstat, pval = ttest_rel(usage_data["pre_usage"], usage_data["Post_usage_1month"])

```

```python
# Degrees of freedom (assuming equal sample sizes)
df = len(usage_data["pre_usage"]) - 1

```

```python
# Critical value (alpha = 0.05, two-tailed test)
alpha = 0.05
critical_value = t.ppf(1 - alpha / 2, df)  # One-tailed critical value for each tail

```

```python
# Print results
print(f"\nT-statistic: {tstat:.2f}, p-value: {pval:.4f}")
print(f"Critical value (alpha = {alpha}, two-tailed): ± {critical_value:.2f}")
```

```python

# Interpretation
if pval < alpha:
    print("Reject null hypothesis: There's a significant difference in usage before and after the campaign (p < 0.05).")
else:
    print("Fail to reject null hypothesis: The difference in usage before and after the campaign might be due to chance (p >= 0.05).")

if abs(tstat) > critical_value:
    print("The difference is statistically significant at the 5% level (|t| > critical value).")
else:
    print("The difference is statistically significant (|t| <= critical value).")
```
