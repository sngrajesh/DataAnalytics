```python
from pulp import *
```

```python
# Create the problem variable to contain the problem data
model = LpProblem("CarProductionProblem", LpMaximize)
```

```python
# Create 2 variables for Cars A and B
# Prarmeters : Name, Lower Limit, Upper Limit, Dta Dype
car_A = LpVariable("Car_A", 0, None, LpInteger)
car_B = LpVariable("Car_B", 0, None, LpInteger)
```

```python
# Create Maximize objective function (Profit)
model += 30000 * car_A + 45000 * car_B, "Profit"
```

```python
# Create three constraints, ensuring unique names
model += 3*car_A + 4*car_B <= 30 , "Materical_Constraint_1"     # Constraint 1
model += 5*car_A + 6*car_B <= 60,  "Materical_Constraint_2"     # Constraint 2
model += 1.5*car_A + 3*car_B <= 21,"Labor_Constraint"           # Constraint 3
```

```python
# Optional constraints - Minimum production of Car A ( Commmented Out)
model += car_A >= 10, "Minimum A Production"
```

```python
# The problem solved using PUlp'S CHOICE OF SOLVER
model.solve()
```

```python
# Each of the variables is printed with its resolved optimum value
for v in model.variables():
    print(v.name, "=", v.varValue)
```

```python
# AdditionalInformation (optional)
print("Total Revenue = ", value(model.objective))
```
