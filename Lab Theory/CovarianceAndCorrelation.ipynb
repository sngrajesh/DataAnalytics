```python
print('Correlation of Height and Weight is: ')
print(df[['Height','Weight']].corr())
print('Correlation of Male Height and Weight is: ')
print(df[df['Gender']== 'Male'][['Height','Weight']].corr())
print('Correlation of Female Height and Weight is: ')
print(df[df['Gender']== 'Female'][['Height','Weight']].corr())
```

```python
print('Covariance of Height and Weight is: ')
print(df[['Height','Weight']].cov())
print('Covariance of Male Height and Weight is: ')
print(df[df['Gender']== 'Male'][['Height','Weight']].cov())
print('Covariance of Female Height and Weight is: ')
print(df[df['Gender']== 'Female'][['Height','Weight']].cov())
```
