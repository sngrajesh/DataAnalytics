```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Part-1

```python
def roll_dice():
    # Simulate the rollingn dice twoce, minimum number we get is 1, max is 6
    # Add the results of the two simulations, so possible values will be:
    # (1,1),(2,2,)...(6,6) ... sum will be between 2 and 12
    return np.sum(np.random.randint(1, 7, 2))

s = pd.Series([roll_dice() for i in range(1000)])

# Use displot instead of distplot and 'bins' instead of 'bin'
sns.displot(s, bins=12)
plt.show()
```

### Part - 2

```python
# Someone approaches us saying I will give you 5 dollars if you get 7 and take 1 dollar if you get a number other than 7
# How do we know whart will happen ?
# Our iwb "Monte Carlo Simulation" like function

def monte_carlo_simulation(runs=1000):
    results = np.zeros(2)   # An array, results[0] and results[1] initialize to two zeros
    for i in range(runs):
        if roll_dice() == 7:
            results[0] += 1
        else:
            results[1] += 1
    return results

# Test 3 times and calculate hoemuch you will win versus lose
result = monte_carlo_simulation(1000)
print(result)
print('Total :', result[0]*7 - result[1]*1)
result = monte_carlo_simulation(1000)
print(result)
print('Total :', result[0]*7 - result[1]*1)
result = monte_carlo_simulation(1000)
print(result)
print('Total :', result[0]*7 - result[1]*1)
```

# Part - 3

```python
# Now do it 1000 times: Take some time
results = np.zeros((1000,2))
for i in range(1000):
    results[i] = monte_carlo_simulation()[0]

print(results)

# Let us plot it
fig, ax = plt.subplots()
ax.hist(results, bins=12)
plt.show()


# Our wins/lose
print(result.mean())                # General mean
print(result.mean()*5)              # What we will get on n average
print(result.mean()*4.75)           # Just a marginal change in win reward - see the impact
print(1000 - result.mean())         # What we will pay on an average
print(result.mean()/1000)           # What is

# Last probability should be close to the theoretical probability of getting
# A 7 when we throw two dice (Why? 1+6 , 2+5, 3+4, 4+3, 5+2, 6+1) = 6 out if 36 state
```
